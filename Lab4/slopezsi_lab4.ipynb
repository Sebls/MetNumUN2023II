**Polinomio en Sympy**

```
from sympy import Symbol
from sympy import div

x = Symbol('x')

p = x**(10) - 8*x**9 - 65*x**8 + 630*x**7 + 483*x**6 - 12684*x**5 + 16505*x**4 + 50870*x**3 - 87484*x**2 - 38808*x + 70560

roots = [1, -1, 2, -2, 3, 4, -5, 6, 7, -7]

for root in roots:
  p, r = div(p,  x-root)
  print("--------------")
  print(f"Monomio: x - ({root})")
  print("Cociente: ", p)
  print("Residuo: ", r)
  print("--------------")

```

```
p = x**(10) - 8*x**9 - 65*x**8 + 630*x**7 + 483*x**6 - 12684*x**5 + 16505*x**4 + 50870*x**3 - 87484*x**2 - 38808*x + 70560

nonroots = [10, 11, 20, 21, 25]

for nonroot in nonroots:
  p, r = div(p,  x-nonroot)
  print("--------------")
  print(f"Monomio: x - ({nonroot})")
  print("Cociente: ", p)
  print("Residuo: ", r)
  print("--------------")
```

## SymPy’s Symbolic and Numercical Complex Evaluations
- https://docs.sympy.org/latest/modules/evalf.html](https://)
- x = x1 + I*x2,y = y1 + I*y2, z = z1 + I*z2, x * y * z

image.png
image.png

```
from sympy import symbols, I, N, pi, expand
x1, x2, x3, y1, y2, z1, z2 = symbols("x1 x2 x3 y1 y2 z1 z2", real=True)
x = -x1*x2+I*(x3)**2
y = (y1)**3-I*(y2)
z = I*z1-(z2)**4

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))
w=1/(pi + I)
print('w=',w)
w=N(w)
print('w=',w)
```
